```python
#Importing Libraries

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

image = cv2.imread('images.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
#Creating Gray Image
grayimage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(grayimage)
```
